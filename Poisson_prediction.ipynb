```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1317 = pd.read_csv('13_17_matches.csv')
df1317
```

```python
df1824 = pd.read_csv('18_24_matches.csv')
df1824
```

```python
df1824 = df1824[['Unnamed: 0','date', 'time', 'comp', 'round', 'day', 'venue', 'gf',
       'ga', 'opponent', 'poss', 'sh', 'save%', 'season', 'team']]
```

```python
df = pd.concat([df1824,df1317], axis = 0).reset_index(drop=True)
df
```

```python
df=df.drop('Unnamed: 0', axis = 1)
df.info()
```

```python
df['total_goals'] = df.gf + df.ga
df.describe()
```

```python
df.loc[[np.argmax(df.total_goals)]]
```

```python
# most_probable home score
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize = (10,5))
ax = fig.gca()
plt.hist(df.gf,bins=30,edgecolor='white')
plt.xticks(range(18));
```

```python
#most probable total goals in a match
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,5))
ax = fig.gca()
plt.hist(df.total_goals,bins=30,edgecolor='black')
plt.xticks(range(20));
```
